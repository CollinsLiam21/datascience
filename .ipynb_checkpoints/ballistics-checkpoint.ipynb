```python
from math import radians, sin, pi
from numpy import linspace
import matplotlib.pyplot as plt
```

```python
def rangeformula(v, theta):
    """
    Inputs:
        v - velocity in m/s 
        theta - launch angle with respect to horizon in radians
    Returns:
        Impact distance in m
    """
    return (v**2/9.81 * sin(2*radians(theta)))
```

```python
rangeformula(100, 0.5)
```

```python
angledegrees = linspace(0,90,num = 200)
```

```python
rangefromdegrees = []
for n in angledegrees:
    rangefromdegrees.append(rangeformula(100,n))
```

```python
plt.plot(angledegrees, rangefromdegrees)
plt.title('Range as a function of angle at 100 m/s')
plt.xlabel('Angle (degrees)')
plt.ylabel('Range (meters)')
plt.show()
```

```python
rangeforangle100 = [rangeformula(100,angle) for angle in angledegrees]
rangeforangle200 = [rangeformula(200,angle) for angle in angledegrees]
```

```python
plt.plot(angledegrees, rangeforangle100, label = "100 m/s")
plt.legend()
plt.plot(angledegrees, rangeforangle200, label = "200 m/s")
plt.legend()
plt.title('Range as a function of angle at various velocities')
plt.xlabel('Angle (degrees)')
plt.ylabel('Range (meters)')
plt.show()
```

# Exercise 1

```python
def rangeformularadians(v, radian):
    """
    Inputs:
        v - velocity in m/s 
        theta - launch angle with respect to horizon in radians
    Returns:
        Impact distance in m
    """
    return (v**2/9.81 * sin(2*radian))
```

```python
angleradians = linspace(0,2*pi,num = 200)
```

```python
rangefromradians = []
for n in angleradians:
    rangefromradians.append(rangeformularadians(100,n))
```

```python
plt.plot(angleradians, rangefromradians)
plt.title('Range as a function of angle at 100 m/s')
plt.xlabel('Angle (radians)')
plt.ylabel('Range (meters)')
plt.show()
```

# Exercise 2

```python
initvelocity = linspace(0,200,num = 200)
```

```python
rangefromvelocity = []
for n in initvelocity:
    rangefromvelocity.append(rangeformula(n,45))
```

```python
plt.plot(initvelocity, rangefromvelocity)
plt.title('Range as a function of initial velocity at angle of 45 degrees')
plt.xlabel('Initial Velocity (m/s)')
plt.ylabel('Range (meters)')
plt.show()
```

# Exercise 3

```python
rangeforvel15 = [rangeformula(v,15) for v in initvelocity]
rangeforvel30 = [rangeformula(v,30) for v in initvelocity]
rangeforvel45 = [rangeformula(v,45) for v in initvelocity]
rangeforvel60 = [rangeformula(v,60) for v in initvelocity]
rangeforvel75 = [rangeformula(v,75) for v in initvelocity]
```

```python
plt.plot(initvelocity, rangeforvel15, label = "15 degrees")
plt.legend()
plt.plot(initvelocity, rangeforvel30, label = "30 degrees")
plt.legend()
plt.plot(initvelocity, rangeforvel45, label = "45 degrees")
plt.legend()
plt.plot(initvelocity, rangeforvel60, label = "60 degrees")
plt.legend()
plt.plot(initvelocity, rangeforvel75, label = "75 degrees")
plt.legend()
plt.title('Range as a function of initial velocity at various angles')
plt.xlabel('Initial Velocity (m/s)')
plt.ylabel('Range (meters)')
plt.show()
```

# Exercise 4

Advantages
- Able to use a list comprehension/for loop to produce a list of range values instead of having to use many cells
- Can alter all of the data by changing one thing in the formula (in the spreadsheet you have to drag down if altering something that's not a variable defined in a different cell)

Disadvantages
- Harder to change one range value or one angle value on the graph
- Graphs take more steps to create (all the commands plt.plot(), plt.legend() etc.)
