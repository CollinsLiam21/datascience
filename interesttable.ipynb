# Lesson 2 - Using DataFrame

Program calculates the growth of an investment over time growing at an annual interest rate of 3.5% using DataFrame. 

Formula Used: 1000(1+0.035)^t

```python
from numpy import radians, sin
from numpy import linspace
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.DataFrame()
df['years'] = range(0,31)
df['value_float'] = 1000*(1 + 0.035)**(df.years)
df['value_str'] = [('$'+str(round(x))) for x in df.value_float]
```

```python
df
```

```python
plt.plot(df.years,df.value_float)
plt.xlabel('Time (years)')
plt.ylabel('Investement Value ($)')
plt.show()
```
